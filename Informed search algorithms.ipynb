### Uniform cost search algorithm

```python
import heapq

def uniform_cost_search(start, goal, cost_dict):
    # Initialize the priority queue (open set) and push the start node with a cost of 0
    open_set = []
    heapq.heappush(open_set, (0, start))
    
    # Dictionary to keep track of the path (where each node came from)
    came_from = {}
    
    # Dictionary to store the cost from the start node to each node
    g_score = {city: float('inf') for city in cost_dict}
    g_score[start] = 0
    
    # Set to keep track of visited nodes
    visited = set()

    # Main loop for the UCS algorithm
    while open_set:
        # Pop the node with the lowest cumulative cost (g_score)
        current_cost, current = heapq.heappop(open_set)

        # Add current node to the visited set
        visited.add(current)

        # If the goal node is reached, reconstruct the path
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            path.reverse()  # Reverse the path to get it from start to goal
            print("Visited nodes:", visited)
            return path

        # Explore all the neighbors of the current node
        for neighbor, cost in cost_dict.get(current, {}).items():
            # Calculate the cumulative cost to reach the neighbor
            tentative_g_score = g_score[current] + cost
            
            # If this path to neighbor is cheaper, update the path and g_score
            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                heapq.heappush(open_set, (tentative_g_score, neighbor))

    # If the goal is not reachable, print visited nodes and return None
    print("Visited nodes:", visited)
    return None

# From and to
start_city = "Addis Ababa"
goal_city = "Moyale"


# Find the path using UCS
path = uniform_cost_search(start_city, goal_city, cost_dict)
print("Path found:", path)
```

### Greedy search

```python
import heapq

def greedy_best_first_search(start, goal, cost_dict, heuristics):
    # Initialize the priority queue (open set) with the start node and its heuristic estimate
    open_set = []
    heapq.heappush(open_set, (heuristics[start], start))  # Prioritize by heuristic estimate only
    
    # Dictionary to track the previous node in the optimal path to each node
    came_from = {}
    
    # Set to keep track of visited nodes
    visited = set()
    visited.add(start)

    while open_set:
        # Pop the node with the smallest heuristic estimate from the priority queue
        _, current = heapq.heappop(open_set)

        # If the goal node is reached, reconstruct the path and return it
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)  # Append the start node to the path
            path.reverse()  # Reverse the path to get it from start to goal
            
            # Print all visited nodes before returning the path
            print("Explored nodes:", visited)
            
            return path

        # Explore neighbors of the current node
        for neighbor, cost in cost_dict.get(current, {}).items():
            if neighbor not in visited:
                # Update the path if this is a better path to the neighbor
                came_from[neighbor] = current
                # Push the neighbor into the priority queue with its heuristic estimate
                heapq.heappush(open_set, (heuristics[neighbor], neighbor))
                # Mark neighbor as visited
                visited.add(neighbor)

    # If the goal is not reachable, return None (no path found)
    print("Explored nodes:", visited)
    return None

# Example usage
start_city = "Addis Ababa"
goal_city = "Moyale"

# Find and print the path using greedy best-first search
path = greedy_best_first_search(start_city, goal_city, cost_dict, heuristics)
print("Path found:", path)
```

### A* search algorithm

```python
import heapq

def a_star_search(start, goal, cost_dict, heuristics):
    # Initialize the priority queue (open set) with the start node and its f_score
    open_set = []
    heapq.heappush(open_set, (0, start))
    
    # Dictionary to keep track of the previous node in the optimal path to each node
    came_from = {}
    
    # Dictionary to store the cost of getting from the start node to each node (g_score)
    g_score = {city: float('inf') for city in heuristics}
    g_score[start] = 0
    
    # Dictionary to store the estimated total cost from start to goal through each node (f_score)
    f_score = {city: float('inf') for city in heuristics}
    f_score[start] = heuristics[start]
    
    # Set to keep track of visited nodes
    visited = set()
    visited.add(start)

    # Main loop for the A* search algorithm
    while open_set:
        # Pop the node with the lowest f_score from the priority queue
        _, current = heapq.heappop(open_set)

        # If the goal node is reached, reconstruct the path and return it
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)  # Append the start node to the path
            path.reverse()  # Reverse the path to get it from start to goal
            
            # Print all visited nodes before returning the path
            print("Explored nodes:", visited)
            
            return path

        # Explore neighbors of the current node
        for neighbor, cost in cost_dict.get(current, {}).items():
            # Calculate the tentative g_score for reaching the neighbor through the current node
            tentative_g_score = g_score[current] + cost
            
            # Update the path if this is a better path to the neighbor
            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score[neighbor] = g_score[neighbor] + heuristics[neighbor]
                # Push the neighbor into the priority queue with its updated f_score
                heapq.heappush(open_set, (f_score[neighbor], neighbor))
                # Add neighbor to visited set
                visited.add(neighbor)

    # If the goal is not reachable, return None (no path found)
    print("Explored nodes:", visited)
    return None

# Example usage
start_city = "Addis Ababa"
goal_city = "Moyale"

# Find and print the path using A* search
path = a_star_search(start_city, goal_city, cost_dict, heuristics)
print("Path found:", path)
```
